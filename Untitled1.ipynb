```python
import numpy as np
import cupy as cp
a = [128, 256, 512, 1024, 2048]

for x in a:
    print(f"\nSize: {x}x{x} (float 32)")

    n = np.random.rand(x, x).astype(np.float32)
    c = cp.array(n)

    print("NumPy:")
    %timeit np.fft.fft2(n)

    print("CuPy:")
    %timeit cp.fft.fft2(c); cp.cuda.Stream.null.synchronize()
```
